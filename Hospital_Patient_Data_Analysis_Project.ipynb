```python
df = df.rename(columns={'Unnamed: 0' : 'index'})
df
# Renamed the column name 'Unnamed: 0' to 'index'
```

```python
df['age'] = df['age']*10
df['income'] = df['income']*10000
# Age was incremented to 10
# Income was incremeted to 1000
```

```python
# We can change age datatype to int because there is no point to make it with decimal point
df['age'] = df['age'].astype(int)
df
```

```python
df.to_csv('NSMES1988-NSMES1988 Processed.csv')
# We saved processed file in NSMES1988-NSMES1988 Processed.csv
```

```python
print(df['visits'].mean()) # Average visit number is 5.8
print(df['visits'].max()) # Max visit number is 89
print(df['visits'].min()) # Min visit number is 0
# Hospital has 5.8 average visits , 89 maximum and 0 minimum visits in average, hospital has low visit rate in average
print()
print(df['age'].mean()) # Average age of a patient is 74
print(df['age'].max()) # Max age of patient is 109
print(df['age'].min()) # Min age of patient is 66
# Average age of patient is 74, min age 66 and max 109, this hospital serves seniors
```

#Data Visualization

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('NSMES1988-NSMES1988 Processed.csv')
df
```

```python
sns.countplot(
    data=df,
    x='insurance',
    palette='pastel',
    hue='insurance'
)

plt.title('Chart 1 : Insurance Count')

plt.xlabel('Insured Yes/No', fontsize=12)
plt.show()

# More than 3 quarter of people insured
# Less than quarter of people not insured
# Outcome : Insurance are preferred to have
```

```python
print(df['insurance'].value_counts())
```

```python
sns.countplot(
    data=df,
    x='married',
    palette='pastel',
    hue='married'
)

plt.title('Chart 2 : Marriage Status')
plt.show()

# Outcome : Married people pays more visit to hospital
```

```python
# Scatter plot to see how number of visits correlated with Health condition
sns.stripplot(data=df, x='health', y='visits', jitter=0.25, alpha=0.7, edgecolor='k')

plt.xlabel('Health')
plt.ylabel('Visits')
plt.title('Chart 3 : Health vs. Visits')
plt.show()

# Bar plot to see how total number of visits correlated with Health condition
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='health', y='visits', estimator=sum)

plt.xlabel('Health')
plt.ylabel('Total Number of Visits')
plt.title('Chart 4 : Total Number of Visits by Health')
plt.show()


# Outcome:
# According to graph we can say highest numbered visits came from people with'poor health'
# Most of the visits belong to people with average health condition
# Less of the visits belong to people with excellent health condition
```

```python
# Counts of age of patients
sns.histplot(df['age'], bins=20, kde=True)
plt.xlabel('Age')
plt.title('Chart 5 : Distribution of Age')
plt.show()

# Health conditions based on age
sns.stripplot(data=df, x='health', y='age', jitter=0.25, alpha=0.7, edgecolor='k')
plt.xlabel('Health')
plt.ylabel('Age')
plt.title('Chart 6 : Health vs. Age')
plt.show()

# Biggest portion of patients are between age 65-70
# Most of the Poor health condition distributed between age 65-85
# Most of the Excellent health condition distributed between age 65-80
# Outcome: There is no significant point shows relation between age and health
```

```python
# Health vs Chronic Conditions
sns.stripplot(data=df, x='health', y='chronic', jitter=0.25, alpha=0.7, edgecolor='k')

plt.xlabel('Health')
plt.ylabel('Chronic Condition')
plt.title('Chart 7 : Health vs. Chronic Condition')
plt.show()

#print(df['chronic'].value_counts())
#print(df['chronic'].sum())

# Most of the patient with Average health has chronic condition between 0-5
# Most of the patient with Poor health has chronic condition between 1-5
# Most of the patient with Excellent health has chronic condition between 0-2
# 85% of the patients has chronic conditions
# Outcome: Most of the patients have chronic conditions
```

```python
# Scatter plot to see how income and private insurance related each other

sns.stripplot(data=df, x='insurance', y='income', jitter=0.25, alpha=0.7, edgecolor='k')
plt.xlabel('Insurance')
plt.ylabel('Income')
plt.title('Chart 8 : Income Distribution on Insurance')
plt.show()

# People with the above average income has No insurance
# People with the below average and low income more likely to have insurance
# Outcome: Income level could be indicater to have insurance
```

```python
# Count plot to see Medicaid count

sns.countplot(
    data=df,
    x='medicaid',
    palette='pastel',
    hue='medicaid'
)
plt.title('Chart 9 : Medicaid Count')
plt.xlabel('Medicaid Yes/No')
plt.show()

# Scatter plot to see how Income and Medicaid related each other
sns.stripplot(data=df, x='medicaid', y='income', jitter=0.25, alpha=0.7, edgecolor='k')
plt.xlabel('Medicaid')
plt.ylabel('Income')
plt.title('Chart 10 : Income Distribution on Medicaid')
plt.show()

print(df['medicaid'].value_counts())



# People with high income less likely to have Medicaid
# Most of the Medicaid people belongs to low income scale
# Most of the people have no Medicaid
# Outcome: There is a relation with income and Medicaid. In genreal patients has less subscription to Medicaid
```

```python
# Scatter plot to see how income and private insurance related each other
plt.scatter(df['income'], df['insurance'], alpha=0.6, edgecolor='k')
plt.xlabel('Income')
plt.ylabel('Insurance')
plt.title('Chart 11 : Income vs Insurance')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()

# Outcome:
# People with the most income has no insurance
# People with the less income more likely to have insurance
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='gender', y='visits', estimator=sum)

plt.xlabel('Gender')
plt.ylabel('Total Number of Visits')
plt.title('Chart 12 : Total Number of Visits by Gender')
plt.show()

# Female patients are more than 50% higher than Male patients
# Outcome : Most of the patients are female
```

```python
#Story

# Scatter plot to see how income and private insurance related each other
plt.scatter(df['income'], df['insurance'], alpha=0.6, edgecolor='k')
plt.xlabel('Income')
plt.ylabel('Insurance')
plt.title('Chart 13 : Income vs Insurance')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(True)
plt.show()


# Box plot to see distribution of income with private insurance
sns.boxplot(data=df, x='insurance', y='income')
plt.xticks([0, 1], ['No', 'Yes'])
plt.xlabel('Insurance')
plt.ylabel('Income')
plt.title('Chart 14 : Income Distribution by Insurance')
plt.show()

sns.stripplot(data=df, x='insurance', y='income', jitter=0.25, alpha=0.7, edgecolor='k')
plt.xlabel('Insurance')
plt.ylabel('Income')
plt.title('Chart 15 : Income Distribution on Insurance')
plt.show()

#Analysis
'''
People with the most income has no insurance and people with the less income more likely to have insurance.
We can think that people with high income no need for insurance , they may be use special insurance or visiting special clinics but low income patients
need to have to have an insurance for their healthcare to prevent high costs of no-insurance scenario.
There are many outliers for both insurance yes and no meaning there is no linear distrubution between income and insurance then, income distribution is wide for having or not having insurance

'''
```
