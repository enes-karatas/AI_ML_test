###Dataset :

Input dataset : FloridaBikeRentals.csv  
Dataset link : https://drive.google.com/file/d/1BAJ8iDpCJdfZSg1QS62RlMiJSs0O8MrG/view

Dataset Columns :

Date: The date when the data was recorded

Rented Bike Count: The number of bikes rented during the given hour
Hour: The hour of the day (0-23)

Temperature(°C): The recorded temperature in Celsius

Humidity(%): The relative humidity percentage

Wind speed (m/s): Wind speed measured in meters per second

Visibility (10m): Visibility recorded in units of 10 meters

Dew point temperature(°C): The dew point temperature in Celsius

Solar Radiation (MJ/m2): The amount of solar radiation received

Rainfall(mm): The recorded rainfall in millimeters

Snowfall (cm): The recorded snowfall in centimeters

Seasons: The season when the data was collected (e.g., Winter, Spring, Summer, Fall)

Holiday: Whether the day was a holiday or not

Functioning Day: Indicates whether the bike rental service was operational on that day

#1. Data Wrangling

```python
# Used try-catch to overcome UnicodeDecodeError error while readung csv file
try:
    df_raw = pd.read_csv('FloridaBikeRentals.csv', encoding='latin-1')
except UnicodeDecodeError:
    df_raw = pd.read_csv('FloridaBikeRentals.csv', encoding='cp1252')

df_raw

df = df_raw.copy()
```

```python
df.sample(30)
print(df['Holiday'].unique())
print(df['Seasons'].unique())
#print(df['Date'].unique())

# Just checking any NaN , Error, or dirty data mixed.
# No nul, NaN , Error, Unkown ... etc found so far, All clean
```

```python
df.drop_duplicates()
df.info()

# No duplicates found
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y' )

# df = df.astype({ 'Rented Bike Count' : 'Int16',
#                 'Hour' : 'Int8',
#                  'Temperature(°C)' : 'float16',
#                  'Humidity(%)' : 'Int16',
#                  'Wind speed (m/s)' : 'float16',
#                  'Visibility (10m)' : 'Int16',
#                  'Dew point temperature(°C)' : 'float16',
#                  'Solar Radiation (MJ/m2)' : 'float16',
#                  'Rainfall(mm)' : 'float16',
#                  'Snowfall (cm)' : 'float16'

# })

df = df.astype({
                 'Temperature(°C)' : 'float32',
                 'Humidity(%)' : 'Int16',
                 'Wind speed (m/s)' : 'float16',
})


# 'Date' datatype casted to Date-Time format
# Some DataTypes are in small but created as -64 format, we can cast it to lower like -16 to -8
# Data types casted from 64 to 16 to reduce memory usage, Memory usage dropped from 958.3+ KB to  829.9+ KB
# No missing data found

df.info()
```

```python
df_cleaned = df.copy()
df_cleaned.info()
```

# 2. Data Processing and Statistical Analysis

```python
df_cleaned.head()
```

```python
df_cleaned['Temperature(°C)'] = df_cleaned['Temperature(°C)'] * 10
df_cleaned.head()

# Temperature columns values multiplied by 10
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_cleaned['Visibility (10m)'] = scaler.fit_transform(df_cleaned[['Visibility (10m)']])

df_cleaned.sample(10)

# Min-Max scaling was applied to 'Visibility (10m)' column
```

```python
df_cleaned[['Temperature(°C)' , 'Humidity(%)', 'Rented Bike Count', 'Visibility (10m)']].describe()

# It gave mean value for Temperature(°C) 'inf' which means infinity, I believe float16 was not enough for this column so I go back and casted it to float32
```

```python
df_raw[['Temperature(°C)' , 'Humidity(%)', 'Rented Bike Count','Visibility (10m)']].describe()
```

```python
# Comparinf to raw data,
# Temperature(°C) data has changed since we applied multiplication
# Visibility data has changed since we applied Min-Max scaling to make it suitable for machine learning models
```

```python
# Categorical Columns : ['Seasons', 'Holiday', 'Functioning Day'] are not suitable for statistical analysis
# We can encode those columns to boolean and then 1 and 0 then we can apply statistical analysis on them
# Columns with dtype int and float are good to go for statistical analysis

# Numerical Columns : ['Rented Bike Count', 'Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)', 'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']
# Categorical Columns : ['Seasons', 'Holiday', 'Functioning Day']
```

# 3. Data Analysis

```python
# Numerical columns and Categorical Columns

numerical_columns = df_cleaned.select_dtypes(include='number').columns.tolist()
categorical_columns = df_cleaned.select_dtypes(include=['object', 'category']).columns.tolist()

print(f'Numerical Columns : {numerical_columns}')
print(f'Categorical Columns : {categorical_columns}')

df_cleaned.dtypes

# We have 10 Numerical 3 categorical columns and also one date.time column
```

```python
df_cleaned.head(10)
```

```python
# Pivotting

avg_rent_bike_count_by_season = pd.pivot_table(df_cleaned, index=['Seasons'],  values='Rented Bike Count', aggfunc='mean')
avg_rent_bike_count_by_season

# Summer season is best season for the business
```

```python
total_rent_bike_count_by_season = pd.pivot_table(df_cleaned, index=['Seasons'],  values='Rented Bike Count', aggfunc='sum')
total_rent_bike_count_by_season
```

```python
avg_rent_bike_count_by_season_func_day = pd.pivot_table(df_cleaned, index=['Seasons','Functioning Day'],  values='Rented Bike Count', aggfunc='mean')
avg_rent_bike_count_by_season_func_day

# In summer and Winter Business operates every day, in autumn and Spring it gives break sometimes
```

```python
avg_rent_bike_count_by_season_holiday = pd.pivot_table(df_cleaned, index=['Seasons','Holiday'],  values='Rented Bike Count', aggfunc='mean')
avg_rent_bike_count_by_season_holiday

# In holidays average rent is lower for No Holiday day in all seasons
```

```python
# Distribution

# Temperature and Rented Bike Count distribution by Hour

df.groupby('Hour')[['Temperature(°C)', 'Rented Bike Count']].mean()

# Highest Rent average is in hour 18&19 and Temperature while 15&14
```

```python
# Seasons and Rented Bike Count distribution on average

df.groupby('Seasons')['Rented Bike Count'].mean()
#df.groupby('Seasons')['Rented Bike Count'].sum()

# Most active season is summer, Winter is very slow
```

```python
# Seasons and Rented Bike Count distribution by total
sns.barplot(data=df_cleaned, x='Seasons', y='Rented Bike Count', estimator=sum)

# Most active season is summer, Winter is very slow
```

```python
# Encoding

df_clean_encoded = pd.get_dummies(df_cleaned)

df_clean_encoded.head(1)

# We've encoded Categorical data as Booleans
```

```python
# Since columns converted to Boolean, I like to convert them to 1 and 0 s in order to make them manipulate ready for statistic operations

boolean_columns = df_clean_encoded.select_dtypes(include='bool').columns
df_clean_encoded[boolean_columns] = df_clean_encoded[boolean_columns].astype(int)

df_clean_encoded
```

#4. Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_cleaned.head(1)
```

```python
#Bar plot for average rentals by Seasons

sns.barplot(data=df_cleaned, x='Seasons', y='Rented Bike Count', estimator=np.mean)

plt.xlabel('Seasons')
plt.ylabel('Average Rented Bike Count ')
plt.title('Average Bike Rentals over Seasons')
plt.show()

# Summer is the best season for bike rentals, Autumn and Spring also good
# Winter is far worst season for bike rentals compare to other seasons
```

```python
# Line plot showing hourly rentals throughout the day

sns.lineplot(data=df_cleaned, x='Hour', y='Rented Bike Count')

plt.xlabel('Hours')
plt.ylabel('Rented Bike Count')
plt.title('Bike Rentals over Hours in a Day')
plt.show()

# Most busy hours are around 6pm and 7pm
# Less busy hours are around 4-5am
# First peak happens around 8am , second and biggest peak is in around 7pm
```

```python
# Heatmap showing correlation among numerical variables

# Only numerical data can be mapped in heatmap, selecting numeric data below
numerical_df = df_cleaned.select_dtypes(include=np.number)

# Computing correlation matrix
corr_matrix = numerical_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Numeric Datas')
plt.show()

# Biggest correlation with Bike rentals is Temperature and then hour
# There is negative correlation to Humidity, Rain and Snow whic makes sense since humans does not like to stay outside in these weather conditions
```

```python
# Box plot to identify outliers in Temperature and Rented Bike Count


#sns.boxplot(data=df_cleaned, x=[['Temperature(°C)' , 'Rented Bike Count	']])

plt.boxplot( [ df_cleaned['Temperature(°C)'], df_cleaned['Rented Bike Count']], labels=['Temperature(°C)' ,'Rented Bike Count'], patch_artist=True, notch=True  )

# Add labels and title for clarity
plt.ylabel('Rented Bike Count')
plt.title('Boxplot visual of Temperature and Bike Rent ')
plt.show()

# There is no outliers for Temperature data
# There are lots of outliers for 'Rented Bike Count' data between around 2300-3500
```
